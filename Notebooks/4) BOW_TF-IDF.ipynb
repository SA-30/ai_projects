## BOW 
(*bag of words*)

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer()
count_vec_fit = count_vec.fit_transform(data)
```

```python
bag_of_words = pd.DataFrame(count_vec_fit.toarray(), columns=count_vec.get_feature_names_out())
bag_of_words
```

## TF IDF 
*term frequency inverse document frequency*

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vec = TfidfVectorizer()
tfidf_vec_fit = tfidf_vec.fit_transform(data)
```

```python
tfidf = pd.DataFrame(tfidf_vec_fit.toarray(), columns=tfidf_vec.get_feature_names_out())
tfidf
```
